```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
files = os.listdir('./data/')
print files
faces_path = []
labels = []
for f in files:
    try:
        ims = os.listdir('./data/' + f)
        print f, ims
        for px in ims:
            faces_path.append('./data/' + f + '/' + px)
            labels.append(f)
    except:
        continue

print '\n\n', labels, faces_path
```

```python
imgs = []
for fx in faces_path:
    im = cv2.imread(fx)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    imgs.append(im)
print len(imgs)
```

```python
for ix in range(len(imgs)):
    plt.figure(ix)
    print imgs[ix].shape
    plt.imshow(imgs[ix])

plt.show()
```

```python
all_faces = []

for ix in range(0,5):
    plt.figure(ix)
    im = imgs[ix][100:600,600:1000,:]              #Cropping images for faces
    img = cv2.resize(im, (100,100))                #Resizing to 100x100 for standardization of all photos
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)    #Converting RGB to Grayscale to ease comparison by a factor of 3
    plt.imshow(img, cmap='gray')
    all_faces.append(img)
    
for ix in range(5,10):
    plt.figure(ix)
    im = imgs[ix][100:300,200:380,:]
    #print im2.shape
    img = cv2.resize(im, (100,100))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    #print img2.shape
    plt.imshow(img, cmap='gray')
    all_faces.append(img)

for ix in range(10, 15):
    plt.figure(ix)
    im = imgs[ix][50:300,150:400,:]
    #print im3.shape
    img = cv2.resize(im , (100,100))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    #print img3.shape
    plt.imshow(img, cmap='gray')
    all_faces.append(img)

plt.show()
```

```python
lab = np.asarray(labels)
print lab.shape

np.save('./data/my_face_labels', lab)        #saving as numpy files
np.save('./data/my_face_data', all_faces)
```
